**Task1 : Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**

### Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read CSV and display few line

```python
df = pd.read_csv("world_population.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

### Using Seaborn Library

```python
sns.barplot(x='Continent', y='2022 Population', data=df)
```

```python
sns.histplot(data = df, x = "Continent", kde = True)
```

### Using Matplotlib Library

```python
Continent = df['Continent']
Population= df['2022 Population']

plt.bar(Continent,Population,width=0.50,color = "green")
plt.show()
```

```python
plt.hist(Continent, bins=15, rwidth=1000,color="Red")
```
